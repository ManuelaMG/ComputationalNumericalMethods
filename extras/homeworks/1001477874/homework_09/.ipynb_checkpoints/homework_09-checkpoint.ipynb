## Last homework
Solve the same equations using `scipy.integrate.odeint` and `scipy.integrate.solve_ivp`

$$
\begin{align*} 
\frac{dx}{dt}&=\sigma (y-x) \\
\frac{dy}{dt}&=x(\rho-z) -y \\
\frac{dz}{dt}&=xy-\beta z 
\end{align*} 
$$

for the parameters:

* $\sigma =10$
* $\rho=28$
* $\beta=8/3$

lets set intial conditions:

* x(0)=1.0
* y(0)=1.0
* z(0)=1.0

Compare the results. Seeing anything strange?

```python
import numpy as np
from scipy.integrate import odeint, solve_ivp

def funcion(t, state, sigma, betha, rho):
    x, y, z = state #no estoy seguro por que se usa este State
     
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - betha * z
     
    return [dx, dy, dz]

sigma = 10
rho = 28
betha = 8/3
p = (sigma, betha, rho)
_0 = [1.0, 1.0, 1.0]

t_span = (0.0, 10.0)
t_eval = np.linspace(0.0, 10.0, 100)

solucion = odeint(funcion, y0, t_eval, p, tfirst=True) # tfirst bool, optional: If True, the first two arguments of func (and Dfun, if given) must t, y instead of the default y, t.
solucion2 = solve_ivp(funcion, t_span, _0, t_eval=t_eval, args=p)
```

## La diferencia que veo entre los metodos es simplemente como guardan los resultados, porque en valor son lo mismo, odeint guarda las tres dimenciones en diferentes columnas de un arreglo [x , y, z ]  sin guardar el valor del t
## solve_ivp guarda los resultados dimencionales en tres diferentes arreglos [x] , [y] , [z] y guarda el tiempo en otro arreglo [ t ]

```python
print (f"solucion odeint {solucion}, solucion con solve_ivp {solucion2}")
```
